```sql
DROP TABLE IF EXISTS catalogworkflowdemo.bronze.companies;

CREATE TABLE catalogworkflowdemo.bronze.companies AS
SELECT
  company_name,
  founded_date,
  country
FROM read_files(
  'abfss://workflowdemo@pakaru27sa.dfs.core.windows.net/landingflow/companies/',
  format => 'csv',
  header => true
);
```
